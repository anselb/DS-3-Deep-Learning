```python
X_train = []
y_train = []
for array in scaled_data:
    X_train.append(array[:-1])
    y_train.append(array[-1])

X_train = np.array(X_train)
y_train = np.array(y_train)

X_test = []
y_test = []
for array in scaled_test_data:
    X_test.append(array[:-1])
    y_test.append(array[-1])
    
X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
print("Shape of X_train: {}".format(X_train.shape))
print("Shape of X_test: {}".format(X_test.shape))
```

```python
X_train = X_train.reshape(30162, 13).astype('float32')
X_test = X_test.reshape(15063, 13).astype('float32')
```

```python
y_train = keras.utils.to_categorical(y_train, 2)
y_test = keras.utils.to_categorical(y_test, 2)
y_test
```

```python
model = Sequential()
model.add(Dense(13, activation='sigmoid', input_shape=(13,)))
model.add(Dense(6, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, batch_size=64, epochs=20, verbose=1, validation_data=(X_test, y_test))
```
